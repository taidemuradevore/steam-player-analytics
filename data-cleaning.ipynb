Here I'm using suggested json parsing from the dataset owner on Kaggle.

```python
import json

def parse_steam_json(file_path):
    print(f"Loading {file_path}...")
    
    # 1. Load the huge JSON file
    with open(file_path, 'r', encoding='utf-8') as f:
        raw_data = json.load(f)
    
    print(f"Found {len(raw_data)} games. Parsing data...")
    
    cleaned_rows = []
    
    for app_id, game in raw_data.items():
        try:
            owners_str = game.get('estimated_owners', '0 - 0')
            if owners_str and '-' in owners_str:
                parts = owners_str.split('-')
                avg_owners = (int(parts[0].strip()) + int(parts[1].strip())) / 2
            else:
                avg_owners = 0

            genre_str = ", ".join(game.get('genres', [])) if game.get('genres') else ""
            tag_str = ", ".join(str(k) for k in game.get('tags', {}).keys()) if game.get('tags') else ""
            dev_str = ", ".join(game.get('developers', [])) if game.get('developers') else ""

            # 3. Build the Clean Row
            row = {
                'app_id': app_id,
                'name': game.get('name'),
                'release_date': game.get('release_date'),
                'price': game.get('price', 0.0),
                'discount': 0.0,
                'avg_owners': avg_owners,
                'peak_ccu': game.get('peak_ccu', 0),
                'dlc_count': game.get('dlc_count', 0),
                'metacritic_score': game.get('metacritic_score', 0),
                'user_score': game.get('user_score', 0),
                'positive_reviews': game.get('positive', 0),
                'negative_reviews': game.get('negative', 0),
                'average_playtime': game.get('average_playtime_forever', 0),
                'median_playtime': game.get('median_playtime_forever', 0),
                'genres': genre_str,
                'tags': tag_str,
                'developers': dev_str,
                'required_age': game.get('required_age', 0),
                'windows': game.get('windows', False),
                'mac': game.get('mac', False),
                'linux': game.get('linux', False)
            }
            cleaned_rows.append(row)
            
        except Exception as e:
            continue
        
    df = pd.DataFrame(cleaned_rows)
    
    df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
    
    return df
```

```python
path = kagglehub.dataset_download("fronkongames/steam-games-dataset")
json_path = os.path.join(path, 'games.json')
games_df = parse_steam_json(json_path)
```

```python
df = pd.merge(
  players_df,
  games_df,
  left_on='app_id',
  right_on='app_id',
  how='left'
)

print(df.head())
```

More cleaning and creating new columns for future use

```python
df.describe()
```

```python
df['days_since_launch'] = (df['month'] - df['release_date']).dt.days
df['months_since_launch'] = df['days_since_launch'] // 30

df['all_time_peak'] = df.groupby('app_id')['peak_players'].transform('max')

launch_window = df[df['months_since_launch'].isin([0, 1])]
launch = launch_window.groupby('app_id')['peak_players'].max().reset_index()
launch = launch.rename(columns={'peak_players': 'launch_peak'})

df = pd.merge(df, launch, on='app_id', how='left')

df['all_time_launch_ratio'] = df['all_time_peak'] / df['launch_peak']


# Drop rows where we couldn't find a launch date
df_clean = df.dropna(subset=['months_since_launch'])
# Drop games where we couldn't find a launch date/peak
df_clean = df.dropna(subset=['months_since_launch', 'launch_peak'])
df_clean = df_clean[df_clean['months_since_launch'] >= 0]
df_clean = df_clean[df_clean['all_time_peak'] >= 100]
```

```python
print(df_clean.columns)
```

```python
# I don't think this is absolutely necessary, but I'm dropping some columns that I will not use and just clutter the .describe() The peak ccu sounds like a very important column, but the data there is inaccurate
df_clean = df_clean.drop(['name_y', 'discount', 'user_score', 'required_age', 'windows', 'mac', 'linux', 'peak_ccu', 'dlc_count', 'average_playtime', 'median_playtime'], axis=1) 
```

```python
df_clean['is_green_month'] = df_clean['gain'] > 0
consistency = df_clean.groupby('app_id')['is_green_month'].mean()

df_clean = pd.merge(df_clean, consistency.rename('gain_consistency'), on='app_id', how='left')
```
